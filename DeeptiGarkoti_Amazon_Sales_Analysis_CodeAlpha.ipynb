***TASK 3: DATA VISUALIZATION***


top 10 product categories

```python
plt.figure(figsize=(45,50))
df['category'].value_counts().head(10).plot(kind='bar')
plt.xticks(rotation=70)
plt.xlabel("category")
plt.ylabel("number of products")
plt.title("top 10 product categories")
plt.show()
```

Relation btw Actual Price and Discounted Price

```python
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)

df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
plt.figure(figsize=(8,5))
sns.regplot(
    data=df,
    x='actual_price',
    y='discounted_price',
    scatter_kws={'alpha':0.3}
)
plt.xlim(0,5000)
plt.ylim(0,5000)
plt.title("Relation between Actual and Discounted Prices")
plt.show()
```

Rating Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=10)
plt.xlabel("Rating")
plt.title("Product Rating Distribution")
plt.show()
```

Highest Discount Product

```python
top_discount = df.sort_values(
    by='discount_percentage',
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

plt.barh(
    top_discount['product_name'].str[:25],
    top_discount['discount_percentage']
)

plt.xlabel("Discount Percentage")
plt.ylabel("Product Name")
plt.title("Top 10 Highest Discounted Products")

plt.gca().invert_yaxis()

plt.show()
```

**Conclusion**

- Analyzed Amazon sales dataset using Python.
- Identified top product categories.
- Compared actual price and discounted price.
- Analyzed customer rating distribution.
- Found products with highest discounts.
- Gained insights about pricing and product trends.

***TASK 1: Web scraping***

import library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

website data

```python
url = "https://books.toscrape.com/"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

extract data

```python
books = soup.find_all("h3")

book_names = []

for book in books:
    book_names.append(book.a["title"])

book_names[:10]
```

convert into DataFrame

```python
scraped_data = pd.DataFrame({
    "Book Name": book_names
})

scraped_data.head()
```

save file

```python
scraped_data.to_csv("scraped_books.csv", index=False)
```

**Conclusion**

In this web scraping task, Learned how to collect data from a website using Python.

Used libraries like Requests and BeautifulSoup to extract information and converted the collected data into a structured format using Pandas.

This task helped understand how real-world data can be gathered from websites and prepared for further analysis.
